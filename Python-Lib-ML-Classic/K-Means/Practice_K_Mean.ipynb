**x1=(1   1   4),x2=(2   2   1),x3=(7   5   6),x4=(6   5   5),x5=(1   0   2),x6=(9   8   7),x6=(4   5   7),x7=(2   1   2)**

**اعلان استفاده از کتابخانه های پایتون**

```python
import pandas as pd
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
```

```python
#   استفاده از کتابخانه Pandas
MyData=pd.DataFrame(
        [[1,1,4],
        [2,2,1],
        [7,5,6],
        [6,5,5],
        [1,0,2],
        [9,8,7],
        [4,5,7],
        [2,1,2]],columns=['F1', 'F2', 'F3']) # ساخت دیتافریم از داده ها
```

```python
# استفاده از کتابخانه  matplotlib
# داده ها دارای ۳ فیچر هستند می توان به صورت سه بعدی رسم نمود
fig = plt.figure() 
ax = plt.axes(projection='3d')
ax = plt.axes(projection='3d')
ax.scatter3D(MyData['F1'], MyData['F2'],MyData['F3']); # رسم نقاط در نمودار سه بعدی
```

```python
# استفاده از تابع kmean  از کتابخانه   sklearn
k=2  # تعداد خوشه هایی که می خواهیم دیتا به آن تعداد خوشه بندی شود
km = KMeans(2)  # فراخوانی تابع کی مین که با الگوریتم کی مین را اجرا می کند
clusts = km.fit_predict(MyData)   # بارگزاری تابع و اجرای الگوریتم و بدست آوردن نتایج خوشه بندی
```

```python
MyData['clusts']=clusts  # اضافه کردن لیبل های خوشه بندی به داده ها
print(MyData) # نمایش خروجی دیتا فریم با نتایج خوشه بندی
```

```python
# رسم نتایج خوشه بندی با رنگ بندی متناظر در نمودار سه بعدی
fig = plt.figure()
ax = plt.axes(projection='3d')
ax = plt.axes(projection='3d')
ax.scatter3D(MyData['F1'], MyData['F2'],MyData['F3'],c=MyData['clusts']);
```
